```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
test=data_set['price']
conv_dates=[1 if values == 2014 else 0 for values in data_set.date]
data_set['date']=conv_dates
train=data_set.drop(['id','price'],axis=1)
```

```python
data_set.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=0.10,random_state=2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
reg_test = reg.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
lr=DecisionTreeClassifier()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
dt_test = lr.score(x_test,y_test)
```

```python
lr.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
lr1=RandomForestRegressor()
lr1.fit(x_train,y_train)
```

```python
lr1.score(x_train,y_train)
```

```python
lr1.predict(x_test)
```

```python
rf_test = lr1.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
dtr_test=model.score(x_test,y_test)
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test)
```

```python
gbr_test = clf.score(x_test,y_test)
```

```python
print("Test accuracy of different algorithms for House price prediction:" )
print('\nLinear Regression:',reg.score(x_test,y_test))
print('\nRandom Forest Regressor:',lr1.score(x_test,y_test))
print('\nDecision Tree Regressor:',model.score(x_test,y_test))
print('\nGradient Boosting Regressor:',gbr_test)
```

```python
print("Train accuracy of different algorithms for House price prediction:" )
print('\nLinear Regression:',reg.score(x_train,y_train))
print('\nRandom Forest Regressor:',lr1.score(x_train,y_train))
print('\nDecision Tree Regressor:',model.score(x_train,y_train))
print('\nGradient Boosting Regressor:',clf.score(x_train,y_train))
```
